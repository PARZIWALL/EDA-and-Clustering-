```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import scipy.stats as stats
# Load the datasets
customers_df = pd.read_csv('Customers.csv')
products_df = pd.read_csv('Products.csv')
transactions_df = pd.read_csv('Transactions.csv')

def calculate_comprehensive_metrics(customers, products, transactions):
    # Merge datasets
    full_data = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')

    # Convert 'TransactionDate' to datetime before using .dt accessor
    full_data['TransactionDate'] = pd.to_datetime(full_data['TransactionDate'])

    # 1. Advanced Customer Segmentation
    def rfm_segmentation(data):
        # Recency: Days since last purchase
        today = data['TransactionDate'].max()
        recency = data.groupby('CustomerID')['TransactionDate'].max()
        recency_days = (today - recency).dt.days

        # Frequency: Total number of transactions
        frequency = data.groupby('CustomerID')['TransactionID'].nunique()

        # Monetary: Total spend
        monetary = data.groupby('CustomerID')['TotalValue'].sum()

        # Normalize and score
        r_score = pd.qcut(recency_days, q=4, labels=[4,3,2,1])
        f_score = pd.qcut(frequency, q=4, labels=[1,2,3,4])
        m_score = pd.qcut(monetary, q=4, labels=[1,2,3,4])

        # Combine scores
        rfm_segment = r_score.astype(str) + f_score.astype(str) + m_score.astype(str)
        return rfm_segment

    # 2. Seasonality and Trend Analysis
    monthly_revenue = full_data.groupby(full_data['TransactionDate'].dt.to_period('M'))['TotalValue'].agg(['sum', 'mean', 'count'])

    # 3. Product Performance
    product_performance = full_data.groupby('ProductID').agg({
        'TotalValue': ['sum', 'mean'],
        'Quantity': 'sum',
        'CustomerID': 'nunique'
    })

    # 4. Category Analysis
    category_analysis = full_data.groupby('Category').agg({
        'TotalValue': ['sum', 'mean'],
        'Quantity': 'sum',
        'ProductID': 'nunique',
        'CustomerID': 'nunique'
    })

    # 5. Regional Performance
    regional_performance = full_data.groupby('Region').agg({
        'TotalValue': ['sum', 'mean', 'count'],
        'CustomerID': 'nunique',
        'TransactionID': 'count'
    })

    # 6. Statistical Tests
    def calculate_statistical_significance(data, column):
        categories = data['Category'].unique()
        category_values = [data[data['Category'] == cat][column] for cat in categories]
        f_statistic, p_value = stats.f_oneway(*category_values)
        return f_statistic, p_value

    # Prepare comprehensive results
    comprehensive_metrics = {
        'rfm_segments': rfm_segmentation(full_data),
        'monthly_revenue': monthly_revenue,
        'product_performance': product_performance,
        'category_analysis': category_analysis,
        'regional_performance': regional_performance,
    }

    return comprehensive_metrics

def generate_eda_insights(metrics):
    insights = []

    # 1. Customer Segmentation Insight
    rfm_counts = metrics['rfm_segments'].value_counts()
    top_segment = rfm_counts.index[0]
    insights.append(f"Customer Segmentation: Most prevalent RFM segment is '{top_segment}', representing {rfm_counts[top_segment]/len(metrics['rfm_segments'])*100:.1f}% of customers. This segment suggests a specific behavioral pattern in purchasing frequency and recency.")

    # 2. Revenue Trend Insight
    revenue_trend = metrics['monthly_revenue']
    max_revenue_month = revenue_trend['sum'].idxmax()
    min_revenue_month = revenue_trend['sum'].idxmin()
    insights.append(f"Revenue Seasonality: Highest revenue month {max_revenue_month} generated ${revenue_trend.loc[max_revenue_month, 'sum']:,.2f}, while lowest month {min_revenue_month} produced ${revenue_trend.loc[min_revenue_month, 'sum']:,.2f}. Indicates significant monthly revenue variability.")

    # 3. Product Performance Insight
    top_products = metrics['product_performance'].sort_values(('TotalValue', 'sum'), ascending=False).head(3)
    insights.append(f"Product Performance: Top 3 products contribute significantly to revenue. Highest-performing product generates ${top_products.iloc[0][('TotalValue', 'sum')]:,.2f}, attracting {top_products.iloc[0][('CustomerID', 'nunique')]} unique customers.")

    # 4. Category Analysis Insight
    category_analysis = metrics['category_analysis']
    top_category = category_analysis['TotalValue']['sum'].idxmax()
    insights.append(f"Category Dominance: {top_category} category leads with ${category_analysis.loc[top_category, ('TotalValue', 'sum')]:,.2f} total revenue. This category attracts {category_analysis.loc[top_category, ('CustomerID', 'nunique')]} unique customers across {category_analysis.loc[top_category, ('ProductID', 'nunique')]} unique products.")

    # 5. Regional Performance Insight
    regional_performance = metrics['regional_performance']
    top_region = regional_performance['TotalValue']['sum'].idxmax()
    insights.append(f"Regional Dynamics: {top_region} region emerges as top performer, generating ${regional_performance.loc[top_region, ('TotalValue', 'sum')]:,.2f} with an average transaction value of ${regional_performance.loc[top_region, ('TotalValue', 'mean')]:,.2f}.")

    # 6. Purchase Behavior Insight
    transaction_frequency = metrics['monthly_revenue']['count']
    avg_monthly_transactions = transaction_frequency.mean()
    transaction_volatility = transaction_frequency.std()
    insights.append(f"Transaction Patterns: Average monthly transactions hover around {avg_monthly_transactions:.0f}, with a standard deviation of {transaction_volatility:.0f}, indicating moderate variability in monthly purchase frequency.")

    return insights
```

```python
# Add these lines before passing dataframes to the function
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])

# Then run the metrics calculation
metrics = calculate_comprehensive_metrics(customers_df, products_df, transactions_df)
insights = generate_eda_insights(metrics)
```
